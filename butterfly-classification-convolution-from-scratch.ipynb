```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/kaggle/input/butterfly-image-classification/Training_set.csv')
test_data = pd.read_csv('/kaggle/input/butterfly-image-classification/Testing_set.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data.label = le.fit_transform(data.label)
```

```python
directory = '/kaggle/input/butterfly-image-classification/train'

def read_images(directory):
    images = []
    labels = []
    
    for _, file in data.iterrows():
        if file[0].endswith(".jpg") or file[0].endswith(".png"):
            image = plt.imread(os.path.join(directory, file[0]))
            images.append(image/255.0)
            labels.append(file[1])
    
    return images, labels

images, labels = read_images(directory)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
images = np.array(images)
labels = np.array(labels)
```

```python
# Define the CNN model
model = keras.Sequential()
# model.add(keras.layers.Conv2D(256, kernel_size=(3, 3), activation='relu'))
# model.add(keras.layers.MaxPooling2D(pool_size=(2, 2)))
model.add(keras.layers.Conv2D(128, kernel_size=(3, 3), activation='relu'))
model.add(keras.layers.MaxPooling2D(pool_size=(2, 2)))
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(128, activation='relu'))
model.add(keras.layers.Dense(75, activation='softmax'))

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(images, labels, epochs=6, batch_size=32)
```

```python
directory = '/kaggle/input/butterfly-image-classification/test'

def read_images(directory):
    images = []
#     labels = []
    
    for _, file in test_data.iterrows():
        if file[0].endswith(".jpg") or file[0].endswith(".png"):
            image = plt.imread(os.path.join(directory, file[0]))
            images.append(image/255.0)
#             labels.append(file[1])
    
    return images

images = read_images(directory)
images = np.array(images)
```

```python
predictions = model.predict(images)
```

```python
actual_preds = np.argmax(predictions, axis = 1)
```
